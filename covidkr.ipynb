```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
import sys

# Optional: Use display if running in a Jupyter notebook
try:
    from IPython.display import display
    use_display = True
except ImportError:
    use_display = False

# URL of the raw data file on GitHub
url = "https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv"

# Download the data
try:
    response = requests.get(url)
    response.raise_for_status()
    with open("owid-covid-data.csv", "wb") as f:
        f.write(response.content)
    print("Data downloaded successfully.")
except requests.exceptions.RequestException as e:
    print(f"Error downloading data: {e}")
    sys.exit()

# Load the data into a pandas DataFrame
try:
    df = pd.read_csv("owid-covid-data.csv", parse_dates=['date'])
except FileNotFoundError:
    print("Error: The downloaded file 'owid-covid-data.csv' was not found.")
    sys.exit()

# Filter for South Korea
south_korea_df = df[df['location'] == 'South Korea'].copy()

if south_korea_df.empty:
    print("No data found for South Korea.")
    sys.exit()
else:
    print("South Korea COVID-19 Data")
    if use_display:
        display(south_korea_df.head())
    else:
        print(south_korea_df.head())
    print("\nSummary Statistics:")
    if use_display:
        display(south_korea_df.describe())
    else:
        print(south_korea_df.describe())

    # Plot daily new cases
    if 'new_cases' in south_korea_df.columns:
        plt.figure(figsize=(12, 6))
        plt.plot(south_korea_df['date'], south_korea_df['new_cases'], label='Daily New Cases')
        plt.xlabel('Date')
        plt.ylabel('New Cases')
        plt.title('Daily New COVID-19 Cases in South Korea')
        plt.legend()
        plt.grid(True)
        plt.show()
    else:
        print("Coluna 'new_cases' não encontrada.")

    # Plot cumulative deaths
    if 'total_deaths' in south_korea_df.columns:
        plt.figure(figsize=(12, 6))
        plt.plot(south_korea_df['date'], south_korea_df['total_deaths'], label='Total Deaths', color='red')
        plt.xlabel('Date')
        plt.ylabel('Total Deaths')
        plt.title('Total COVID-19 Deaths in South Korea')
        plt.legend()
        plt.grid(True)
        plt.show()
    else:
        print("Coluna 'total_deaths' não encontrada.")

    # Calculate and print the latest mortality rate
    last_valid_data = south_korea_df.dropna(subset=['total_cases', 'total_deaths'])
    if not last_valid_data.empty:
        latest_data = last_valid_data.iloc[-1]
        if latest_data['total_cases'] > 0:
            mortality_rate = (latest_data['total_deaths'] / latest_data['total_cases']) * 100
            print(f"\nLatest Mortality Rate in South Korea: {mortality_rate:.2f}%")
        else:
            print("\nCannot calculate mortality rate as total cases are zero.")
    else:
        print("\nDados insuficientes para calcular a taxa de mortalidade.")
```
